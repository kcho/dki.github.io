## Load library

```python
import os
import pandas as pd
import re
from rpy2.robjects.lib import ggplot2
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from ggplot import *

pd.set_option('max_rows',10000)
pd.set_option('max_columns',10000)
pd.options.display.mpl_style = 'default'

#stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

## Processed subjects as a list

```python
subjects = [x for x in os.listdir('/Volumes/CCNC_3T/KangIk/2014_05_DKI_project') if re.match('(CHR|FEP|NOR|\d)',x)]
```

## Make a groupMap for group assignment

```python
groupMap={}
for subject in subjects:
    if subject.startswith('FEP') or subject.startswith('CHR'):
        groupMap[subject]=subject[:3]
    elif subject.startswith('NOR') :
        groupMap[subject]='HCs'
    else:
        if re.search('(-|_)T(-|_)',subject):
            groupMap[subject]='SPR'
        elif re.search('(-|_)C(-|_)',subject):
            groupMap[subject]='SPR_HCs'
        else:
            print subject
```

## Read output from each subject

```python
kurtosisDf = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 'NOR85_CJM',
                                 'DKI',
                                 'mean.txt'),
                    sep='\s+',
                    names=['modality','ROI','kurtosis','volume','volume2'])
```

```python
kurtosisDf
```

```python
meanDf = {}
axialDf = {}
radialDf = {}

for subject in subjects:
    diffusivity = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 subject,
                                 'DKI',
                                 'diffusion_mean.txt'),
                    sep='\s+',
                    names=['modality','ROI','diffusivity'])
    
    kurtosisDf = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 subject,
                                 'DKI',
                                 'mean.txt'),
                    sep='\s+',
                    names=['modality','ROI','kurtosis','volume','volume2'])
    
    densityDf = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 subject,
                                 't1Density',
                                 'mean.txt'),
                    sep='\s+',
                    names=['ROI','density'])
        
    diffusivity['subject'] = subject
    diffusivity['ROI']=diffusivity['ROI'].str[4:]
    diffusivity['modality'] = diffusivity['modality'].str[1:]
    
    kurtosisDf['subject'] = subject
    kurtosisDf['ROI']=kurtosisDf['ROI'].str[4:]
    kurtosisDf['modality'] = kurtosisDf['modality'].str[1:]
        
    densityDf['subject'] = subject

    merged_diffusivity_kurtosis = pd.merge(diffusivity,kurtosisDf,how='inner',on=['subject','ROI','modality'])    
    merged = pd.merge(merged_diffusivity_kurtosis,densityDf,how='inner',on=['subject','ROI'])
    newDf = merged[merged['diffusivity']!=0][['subject','ROI','diffusivity','kurtosis','density','volume','modality']]
    newDf['group']=newDf['subject'].map(groupMap)
    
    meanDf[subject] = newDf[newDf['modality']=='mean.nii']
    axialDf[subject] = newDf[newDf['modality']=='ax.nii']
    radialDf[subject] = newDf[newDf['modality']=='rad.nii']
    
meanDf = pd.concat([x for x in meanDf.values()],axis=0)
axialDf = pd.concat([x for x in axialDf.values()],axis=0)
radialDf = pd.concat([x for x in radialDf.values()],axis=0)

meanDf.subject=meanDf.subject.str[:5]
axialDf.subject=axialDf.subject.str[:5]
radialDf.subject=radialDf.subject.str[:5]
```

## Add location to each ROI

```python
locationMap={}
for roi in meanDf.ROI.unique():
    if 'ctx' in roi:
        locationMap[roi]='cortex'
    elif re.search('(Amygdala|Accumbens|Caudate|Hippocampus|Pallidum|Putamen|Thalamus-Proper)',roi):
        locationMap[roi]='subcortex'
    elif 'Ventricle' in roi:
        locationMap[roi]='ventricle'
    elif roi.startswith('lh') or roi.startswith('rh'):
        locationMap[roi]='thalProject'
    elif 'CSF' in roi:
        locationMap[roi]='CSF'
    elif 'Cerebellum-Cortex' in roi:
        locationMap[roi]='cerebellum'
    elif 'Brain-Stem' in roi:
        locationMap[roi]='brain stem'
    elif 'white' in roi.lower():
        locationMap[roi]='white matter'
    else:
        locationMap[roi]='other'
        
```

```python
meanDf['location']=meanDf.ROI.map(locationMap)
axialDf['location']=axialDf.ROI.map(locationMap)
radialDf['location']=radialDf.ROI.map(locationMap)

meanDf = meanDf.reset_index()
axialDf = axialDf.reset_index()
radialDf = radialDf.reset_index()
```

# Scatter plot of mean diffusivity, mean kurtosis, volume, intensity(density) of the ROIs

```python
prac=meanDf.groupby('ROI')
df = prac.mean().reset_index()
df['location']=df.ROI.map(locationMap)

fig,axes = plt.subplots(nrows=4,figsize=(20,60))

for ax,col in zip(axes,['diffusivity','kurtosis','volume','density']):
    ax.plot(df[col],'k.')
    ax.set_title(col.upper())
    ax.set_xlabel('ROIs')
    ax.set_ylabel(col)
    
    #cortex : red
    for where,color in zip(['cortex','subcortex','cerebellum','ventricle','CSF','thalProject'],
                           ['ro','bo','co','go','yo','mo']):
        newDf = df[df.location==where]
        ax.plot(newDf.index,newDf[col],color, label=where)
    
    ax.legend(title='Location')
    #other
    #ax.plot(meanDf[meanDf.location=='other'][col],'k.')
    #ax.set_title(col.upper())
    #ax.set_xlabel('ROIs')
    #ax.set_ylabel(col)
```
