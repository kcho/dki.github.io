## Merge kurtosis data + clinical data

### Merge Df

```python
fep_clinic_kurtosis_df = pd.merge(fepDf,clinicalDf,how='outer',on='subject')
```

## Correlation with kurtosis

```python
roiToCheck=['ctx-lh-pericalcarine.nii.gz',
            'ctx-lh-lingual.nii.gz',
            'ctx-lh-cuneus.nii.gz',
            'ctx-lh-inferiortemporal.nii.gz',
            'ctx-lh-superiorparietal.nii.gz',
            'ctx-lh-middletemporal.nii.gz',
            'ctx-lh-entorhinal.nii.gz',
            'ctx-lh-temporalpole.nii.gz',
            'ctx-lh-paracentral.nii.gz',
            'ctx-lh-parstriangularis.nii.gz',
            'ctx-lh-fusiform.nii.gz',
            'ctx-lh-posteriorcingulate.nii.gz']
```

```python
clinicalDf.columns[9:]
```

```python
for item in clinicalDf.columns[9:]:
    for roi in roiToCheck:
        #print roi
        df = fep_clinic_kurtosis_df[fep_clinic_kurtosis_df.ROI==roi]
        df = df.drop(df[df.ptot.isnull()].index)
        coef,pvalue=stats.pearsonr(df[item],df['kurtosis'])
        if pvalue < 0.05:
            #print pvalue, coef, roi, item
            fig,axes = plt.subplots(figsize=(20,10))
            #axes.plot(df[item],df['kurtosis'],'ko')
            
          
            fit = polyfit(df[item],df['kurtosis'],1)
            fit_fn = poly1d(fit) # fit_fn is now a function which takes in x and returns an estimate for y

            axes.plot(df[item],df['kurtosis'], 'ko', 
                      df[item], fit_fn(df[item]), '--r')
            #axes.set_xlim(left=0)
            #axes.set_ylim(bottom=0)
            axes.set_title(' '.join([roi,item]),fontsize=20)
            axes.set_ylabel('Kurtosis',fontsize=15)
            axes.set_xlabel(item.upper(),fontsize=15)
            axes.set_xbound(lower=df[item].min()-0.5,upper=df[item].max()+0.5)
            #print axes.get_xlim(),axes.get_ylim()
            axes.text((axes.get_xlim()[0]+axes.get_xlim()[1])/2,(axes.get_ylim()[0]+axes.get_ylim()[1])/2,
                      'p-value {0}, coeff {1}'.format(pvalue,coef),
                      style='italic',fontsize=15,
                      bbox={'facecolor':'white',
                            'alpha':0.6,
                            'pad':10})
            #axes.set_xlim(left=df[item].min()-1,right=df[item].min()+1)
            #axes.legend(title='P-value')

            fig.show();
```
