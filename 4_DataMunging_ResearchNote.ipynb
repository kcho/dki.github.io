#Loading the outcomes

- dfs

    - radialDf


```python
subjects= [x for x in os.listdir('/Volumes/CCNC_3T/KangIk/2014_05_DKI_project') if x.startswith('CHR') or x.startswith('NOR') or x.startswith('FEP')]
meanDf = {}
axialDf = {}
radialDf = {}
for subject in subjects:
    try:
        df = pd.read_csv(os.path.join(subject,'DKI','mean.txt'),sep='\s+',names=['ROIs','mean']).reset_index()
        df.columns=['modality','ROIs','mean']
        df['subject'] = subject
        df['group'] = df.subject.map(lambda x:x[:3])

        meanDf[subject] = df[df.modality=='kmean.nii'][['subject','group','ROIs','mean']]
        axialDf[subject] = df[df.modality=='kax.nii'][['subject','group','ROIs','mean']]
        radialDf[subject] = df[df.modality=='krad.nii'][['subject','group','ROIs','mean']]
        
    except:
        pass

meanDf = pd.concat([x for x in meanDf.values()],axis=0)
axialDf = pd.concat([x for x in axialDf.values()],axis=0)
radialDf = pd.concat([x for x in radialDf.values()],axis=0)
```

```python
meanDf.ROIs.drop_duplicates()
```

```python
for roi in meanDf.ROIs.drop_duplicates():
    a = meanDf[meanDf.ROIs==roi]
    display(a.groupby('group').mean().plot(kind='line'))
```
