##4. Extract extra ROIs
Extract all the possible ROIs from freesurfer
```
#!/Users/admin/anaconda/bin/python
import pandas as pd
import os
import re
import argparse


def roiExtraction(subject):

    colourTable = pd.read_csv('/Volumes/CCNC_3T/KangIk/2014_05_DKI_project/FreeSurferColorLUT.txt',sep='\s+')
    colourTable.columns=['roiNumber','labelName','R','G','B','A']
    colourTable

    os.environ["FREESURFER_HOME"] = '/Applications/freesurfer'
    os.environ["SUBJECTS_DIR"] = '{0}'.format(subject)

    print subject

    roiDestination=os.path.join(os.getcwd(),subject,'extraROIs')
    print roiDestination

    try:
        os.mkdir(roiDestination)
    except:
        pass

    #to extract thalamus
    for index,subjDf in colourTable.iterrows():
        #print subjDf['roiNumber']
        #print subjDf['labelName']
        os.system('mri_binarize --i {0}/freesurfer/mri/aparc+aseg.mgz \
        --match {1} \
        --o {0}/extraROIs/{2}.nii.gz'.format(subject,subjDf['roiNumber'],subjDf['labelName']))

if __name__=='__main__':
    argparser = argparse.ArgumentParser(prog='roiExtraction',
            formatter_class=argparse.RawDescriptionHelpFormatter,
            description='''\
                    Extracts extra ROIs
''',epilog="Kevin Cho 2014_06_05")
    argparser.add_argument("--subject","-s",
            help='''
            specify name of a subject folder
            ''')
    args = argparser.parse_args()
    if args.subject:
        roiExtraction(args.subject)
```

##5. Registration of the extra ROIs to diffusion space
### nearest neighbour option is required to make the ROI into binary maps
```
'''
 74     ##=============================================                                                                       
 75     ## Registration of the extra ROIs                                                                                     
 76     ##=============================================                                                                       
 77     echo -ne '\tROI\n'                                                                                                    
 78     for imgs in ${i}/extraROIs/*                                                                                          
 79     do                                                                                                                    
 80         filename=$(basename "$imgs")                                                                                      
 81
 82         flirt -in ${imgs} \                                                                                               
 83             -ref ${i}/DKI/nodif.nii.gz \                                                                                  
 84             -applyxfm -init ${i}/registration/FreesurferToDKI \                                                           
 85             -cost mutualinfo -interp nearestneighbour \                                                                                            
 86             -out ${i}/ROI/dki_${filename}                                                                                 
 87     done       
'''
```
