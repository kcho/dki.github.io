# Correlation between diffusion kurtosis and mean kurtosis

## Load library

```python
from IPython.display import Image
from IPython.display import HTML
import os
import pandas as pd
import re
from rpy2.robjects.lib import ggplot2
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from ggplot import *

pd.set_option('max_rows',10000)
pd.set_option('max_columns',10000)
pd.options.display.mpl_style = 'default'

%load_ext rmagic

%R require(ggplot2)

#stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

## Load data

```python
subjects = [x for x in os.listdir('/Volumes/CCNC_3T/KangIk/2014_05_DKI_project') if x.startswith('NOR')]
```

```python
meanDf = {}
axialDf = {}
radialDf = {}

for subject in subjects:
    df = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 subject,
                                 'DKI',
                                 'diffusion_mean.txt'),
                    sep='\s+',
                    names=['modality','ROI','diffusivity'])
    
    kurtosisDf = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 subject,
                                 'DKI',
                                 'mean.txt'),
                    sep='\s+',
                    names=['modality','ROI','kurtosis','volume1','volume2'])
        
    df['subject'] = subject
    if subject.startswith('FEP'):
        df['group']='FEP'
    else:
        df['group']='HCs'   
    df['ROI']=df['ROI'].str[4:]
    df['modality'] = df['modality'].str[1:]
    
    kurtosisDf['subject'] = subject
    if subject.startswith('FEP'):
        kurtosisDf['group']='FEP'
    else:
        kurtosisDf['group']='HCs'
    kurtosisDf['ROI']=kurtosisDf['ROI'].str[4:]
    kurtosisDf['modality'] = kurtosisDf['modality'].str[1:]
        
    merged = pd.merge(df,kurtosisDf,how='inner',on=['subject','group','ROI','modality'])    
    newDf = merged[merged['diffusivity']!=0][['subject','group','ROI','diffusivity','kurtosis','modality']]
    #print subject[:5], '\t', len(newDf)
    

    meanDf[subject] = newDf[newDf['modality']=='mean.nii']
    axialDf[subject] = newDf[newDf['modality']=='ax.nii']
    radialDf[subject] = newDf[newDf['modality']=='rad.nii']
    
meanDf = pd.concat([x for x in meanDf.values()],axis=0)
axialDf = pd.concat([x for x in axialDf.values()],axis=0)
radialDf = pd.concat([x for x in radialDf.values()],axis=0)
```

# Scatter plot

## Using all the ROIs from Freesurfer

```python
fig,axes = plt.subplots(figsize=(20,10))
meanDf.plot(x='diffusivity',y='kurtosis',kind='scatter',ax=axes);
```

## Cortex

```python
fig,axes = plt.subplots(figsize=(20,10))
meanDf[meanDf.ROI.str.contains('ctx')].plot(x='diffusivity',y='kurtosis',kind='scatter',ax=axes);
```

## Subcortex

```python
fig,axes = plt.subplots(figsize=(20,10))
meanDf[meanDf.ROI.str.contains('Accumbens|Amygdala|Caudate|Hippocampus|Pallidum|Putamen|Thalamus')].plot(x='diffusivity',y='kurtosis',kind='scatter',ax=axes);
```

## CSF

```python
fig,axes = plt.subplots(figsize=(20,10))
meanDf[meanDf.ROI.str.contains('Ventricle|CSF')].plot(x='diffusivity',y='kurtosis',kind='scatter',ax=axes);
```
