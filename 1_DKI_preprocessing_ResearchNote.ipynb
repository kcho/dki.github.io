###Updated sequence of the preprocessing
1. dcm2nii : dicom to nifti conversion
2. Eddy current correction

    'eddy_correct ${i}/*.nii.gz ${i}/dki_eddy.nii.gz 0'
    
3. DKE without smoothing, interpolation nor coregistration
    - CHR : M1
    - FEP : M2
    - NOR : M3
