## Group by other categories

#### Number of medication types

```python
len(df.med.drop_duplicates())
```

#### Random criteria make up

```python
numCategory = np.random.randint(1,5,size=41)
numCategory
```

#### Price tag

```python
numToStrength={1:'high',
               2:'medium',
               3:'low',
               4:'zero'}
```

#### Make into dictionary

```python
map_dictionary = dict(zip(df.med.drop_duplicates().tolist(),[numToStrength[x] for x in numCategory]))
```

#### Groupby the category

```python
gb = df.set_index('med').groupby(map_dictionary)
gb.sum()
```
