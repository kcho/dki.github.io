```python
from IPython.display import Image
from IPython.display import HTML
import os
import pandas as pd
import re
from rpy2.robjects.lib import ggplot2
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from ggplot import *

pd.set_option('max_rows',10000)
pd.set_option('max_columns',10000)
pd.options.display.mpl_style = 'default'

%load_ext rmagic

%R require(ggplot2)

#stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

## Load NOR data

```python
subjects = [x for x in os.listdir('/Volumes/CCNC_3T/KangIk/2014_05_DKI_project') if x.startswith('NOR')]
```

```python
meanDf = {}
axialDf = {}
radialDf = {}

for subject in subjects:
    df = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 subject,
                                 'DKI',
                                 'mean.txt'),
                    sep='\s+',
                    names=['modality','ROI','kurtosis','voxel','volume'])
    
    densityDf = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 subject,
                                 't1Density',
                                 'mean.txt'),
                    sep='\s+',
                    names=['ROI','density'])
        
    df['subject'] = subject
    if subject.startswith('FEP'):
        df['group']='FEP'
    else:
        df['group']='HCs'   
    df['ROI']=df['ROI'].str[4:]
    
    densityDf['subject'] = subject
    if subject.startswith('FEP'):
        densityDf['group']='FEP'
    else:
        densityDf['group']='HCs'   
        
    
    merged = pd.merge(df,densityDf,how='inner',on=['subject','group','ROI'])
    

    newDf = merged[merged['kurtosis']!=0][['subject','group','ROI','kurtosis','voxel','volume','density','modality']]
    #print subject[:5], '\t', len(newDf)
    
    meanDf[subject] = newDf[newDf['modality']=='kmean.nii']
    axialDf[subject] = newDf[newDf['modality']=='kax.nii']
    radialDf[subject] = newDf[newDf['modality']=='krad.nii']
    
meanDf = pd.concat([x for x in meanDf.values()],axis=0)
axialDf = pd.concat([x for x in axialDf.values()],axis=0)
radialDf = pd.concat([x for x in radialDf.values()],axis=0)
```

## Density comparison

### Z-score conversion

```python
zscore = lambda x: (x - x.mean()) / x.std()
```

```python
meanDf['densityZscore'] = (meanDf['density']-meanDf['density'].mean()) / meanDf['density'].std()
meanDf['kurtosisZscore'] = (meanDf['kurtosis']-meanDf['kurtosis'].mean()) / meanDf['kurtosis'].std()
```

### Correlation between density and kurtosis

```python
meanDf[['density','kurtosis']].corr()
```

```python
stats.pearsonr(meanDf[['density']],meanDf[['kurtosis']])
```

```python
fig,axes = plt.subplots(figsize=(20,10))
meanDf.plot(x='density',y='kurtosis',kind='scatter',ax=axes);
```
