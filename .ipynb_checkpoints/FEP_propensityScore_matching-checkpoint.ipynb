## Demographic information description

```python
def lineAndBox(df,column):
    fig, axes = plt.subplots(figsize=(20, 15))

    ax1 = plt.subplot2grid((2,2), (0,0), colspan=2)
    ax2 = plt.subplot2grid((2,2), (1, 0))
    #ax3 = plt.subplot2grid((2,2), (1, 1))

    for key, grp in df[[column,'group']].groupby('group'):
        print key,'subjects :',len(grp)
        ax1.plot(grp[column],label=key)

    ax1.legend(loc='best')

    sns.boxplot(df[column],df.group,ax=ax2)

    ax1.set_title(column)
    fig.show()
```

# Propensity matching summary

```python
matchedDf = pd.read_csv("/Users/admin/Dropbox/project/DKI_2014/demographics/matchedFEP.csv")
matchedDf.group = matchedDf.group.map({0:'NOR',1:'FEP'})
matchedDf = matchedDf[['index','sex','age','IQ','Education','handedness','group']]
matchedDf = matchedDf.set_index('index')
matchedDf = matchedDf.reset_index()
matchedDf.columns = ['subject','sex','age','IQ','Education','handedness','group']
gb = matchedDf.groupby('group')
display(gb.describe().applymap(lambda x: round(x,1)))


fig, axes = plt.subplots(figsize=(20, 15),)

ax1 = plt.subplot2grid((3,3), (0,0))
ax2 = plt.subplot2grid((3,3), (0,1))
ax3 = plt.subplot2grid((3,3), (1,0))
ax4 = plt.subplot2grid((3,3), (1,1))
ax5 = plt.subplot2grid((3,3), (2,0))

ax1.set_title('Age')
sns.boxplot(matchedDf.age,matchedDf.group,ax=ax1)
ax1.set_ylim(bottom=14)

ax2.set_title('Education')
sns.boxplot(matchedDf.Education,matchedDf.group,ax=ax2)
ax2.set_ylim(7,20)

ax3.set_title('Sex')
#cross-tab
subjectCount = pd.crosstab(matchedDf.group, matchedDf.sex)
subjectCount.plot(kind='bar',stacked=False, color=['black','gold'],ax=ax3)
ax3.set_ylabel('Count')


ax4.set_title('IQ')
sns.boxplot(matchedDf.IQ,matchedDf.group,ax=ax4)

ax5.set_title('Handedness')
#cross-tab
subjectCount = pd.crosstab(matchedDf.group, matchedDf.handedness)
subjectCount.plot(kind='bar',stacked=False, color=['black','gold'],ax=ax5)
ax5.set_ylabel('Count')
ax5.set_ylim(top=40)


#ax3 = plt.subplot2grid((2,2), (1, 1))
fig.show()
```

```python
from scipy import stats
```

```python
twoSample = stats.ttest_ind(matchedDf[matchedDf.group=='NOR'][['age']],matchedDf[matchedDf.group=='FEP'][['age']])
print 'The t-statistic is %.3f and the p-value is %.3f.' % twoSample
```

```python
twoSample = stats.ttest_ind(matchedDf[matchedDf.group=='NOR'][['IQ']],matchedDf[matchedDf.group=='FEP'][['IQ']])
print 'The t-statistic is %.3f and the p-value is %.3f.' % twoSample
```

```python
twoSample = stats.ttest_ind(matchedDf[matchedDf.group=='NOR'][['Education']],matchedDf[matchedDf.group=='FEP'][['Education']])
print 'The t-statistic is %.3f and the p-value is %.3f.' % twoSample
```
