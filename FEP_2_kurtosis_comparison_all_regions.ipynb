## Kurtosis collection

```python
subjects = [x for x in os.listdir('/Volumes/CCNC_3T/KangIk/2014_05_DKI_project') if x.startswith('FEP') or x.startswith('NOR')]
```

```python
meanDf = {}
axialDf = {}
radialDf = {}

for subject in subjects:
    df = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 subject,
                                 'DKI',
                                 'mean.txt'),
                    sep='\s+',
                    names=['modality','ROI','kurtosis','voxel','volume'])
    
    densityDf = pd.read_csv(os.path.join(
                                 '/Volumes/CCNC_3T/KangIk/2014_05_DKI_project',
                                 subject,
                                 't1Density',
                                 'mean.txt'),
                    sep='\s+',
                    names=['ROI','density'])
        
    df['subject'] = subject
    if subject.startswith('FEP'):
        df['group']='FEP'
    else:
        df['group']='HCs'   
    df['ROI']=df['ROI'].str[4:]
    
    densityDf['subject'] = subject
    if subject.startswith('FEP'):
        densityDf['group']='FEP'
    else:
        densityDf['group']='HCs'   
        
    
    merged = pd.merge(df,densityDf,how='inner',on=['subject','group','ROI'])
    

    newDf = merged[merged['kurtosis']!=0][['subject','group','ROI','kurtosis','voxel','volume','density','modality']]
    #print subject[:5], '\t', len(newDf)
    
    meanDf[subject] = newDf[newDf['modality']=='kmean.nii']
    axialDf[subject] = newDf[newDf['modality']=='kax.nii']
    radialDf[subject] = newDf[newDf['modality']=='krad.nii']
    
meanDf = pd.concat([x for x in meanDf.values()],axis=0)
axialDf = pd.concat([x for x in axialDf.values()],axis=0)
radialDf = pd.concat([x for x in radialDf.values()],axis=0)
```

```python
meanDf.subject = meanDf.subject.str[:5]
meanDf = meanDf.set_index('subject')
```

# Comparison

```python
def compare(df,ROI,toCompare):
    
#    df = meanDf[meanDf.ROI=='dki_Left-Hippocampus.nii.gz']
#    df.groupby('group').describe().T

    newDf = df[df['ROI']==ROI]
    
    mean = newDf[toCompare]
    GROUP = newDf['group']

    formula = '{0} ~ GROUP'.format(toCompare)

    lm = ols(formula, newDf).fit()

    try:
        output=anova_lm(lm)
#        if output.ix['GROUP','PR(>F)'] < 0.05:
        if output.ix['GROUP','PR(>F)']:
            
            count = pd.crosstab(newDf.group,newDf.index.sum())
 
            
            count.columns=['count']
            print(count)
            print
            
            print '{0} \t\t {1}~GROUP'.format(ROI,toCompare)
            print '========='
            print output
            print '\n\n\n'
            #display(newDf.groupby('group').describe().T.applymap(lambda x: round(x,1)))
     
            name = ROI[4:-7]
     
            fig, axes = plt.subplots(figsize=(8,8))
            ax1 = plt.subplot2grid((1,1), (0,0))
            df = selectMean[selectMean['ROI']==ROI]
            #display(sns.boxplot(df['kurtosis'],df.group,ax=ax1)
            sns.boxplot(df['kurtosis'],df.group,ax=ax1)
            try:
                roiName = re.search('h-(.*)\.nii\.gz',ROI).group(1)
                side = re.search('(rh|lh)-',ROI).group(1)#.map({'rh':'Right','lh':'Left'})
            except:
                search = re.search('(Left|Right)-(.*)\.nii\.gz',ROI)
                roiName = search.group(2)
                side = search.group(1)
                            
            title = roiName[0].upper()+roiName[1:]
            ax1.set_title('{0} {1} {2}'.format(side.upper(),title,toCompare))
            
            show()
            
    except:
        print 'error'
    return newDf
```
